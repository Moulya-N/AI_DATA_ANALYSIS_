```python
# Ques_5.ipynb â€” Validate Data Completeness for Critical Fields

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("transaction_data.csv")

# Define critical fields
critical_fields = ["transaction_id", "amount", "date"]

# Calculate missing percentages
missing_percent = df[critical_fields].isnull().mean() * 100
missing_percent = missing_percent.round(2)

# Display missing percentage
print("Missing Data Percentage for Critical Fields:")
print(missing_percent)

# Visualization
plt.figure(figsize=(8, 5))
missing_percent.plot(kind='bar', color='tomato', edgecolor='black')
plt.title("Missing Data Percentage for Critical Fields")
plt.ylabel("Missing Percentage (%)")
plt.xticks(rotation=0)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Optional: Save result to CSV
missing_percent.to_csv("critical_fields_missing_report.csv", header=["Missing_Percentage"])
```
