```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'height': [160, 165, 170, 175, 180],
    'weight': [55, 60, 65, 70, 75],
    'size': [10, 15, 13, 18, 12]  # Marker sizes
})

# Create scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='height',
    y='weight',
    size='size',
    sizes=(50, 300),  # Adjust the range of marker sizes
    data=data,
    legend=False  # Hide legend if not needed
)

# Customize plot
plt.title('Height vs. Weight with Variable Marker Sizes')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.tight_layout()
plt.show()
```
